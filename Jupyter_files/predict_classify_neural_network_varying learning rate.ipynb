```python
import numpy as np
import matplotlib.pyplot as plt

from neural_network import Network
from layers import fcLayer, classLayer, regLayer 

# training data
#x_train = np.array([[0,0,1,1], [0,1,0,1]]) # m - no of features and n - no of instances
#y_train = np.array([[0,1,1,0]]) 

x_train = np.expand_dims(np.linspace(-1,1,1000),axis=0) #-1 to 0 and 0.75 to 1.5
#x_test  =  #0 to 0.75

m = x_train.shape[1]
np.random.seed(48)
y_train = np.exp(x_train) + 0.02*np.random.randn(1,m)

eta_list = [0.05, 0.1, 0.2, 0.3]
nN = len(eta_list)
for i in range(nN):
    eta = eta_list[i]
    fcnet = Network()
    fcnet.add(fcLayer(1, 10,'sigmoid'))
    fcnet.add(regLayer(10, 1,'linear'))
    fcnet.optimizer(eta=eta, lamda=10, regularizer ='none') #L1
    # train
    [train_loss, test_loss] = fcnet.fit(x_train, y_train, x_train, y_train, batch_size = 8, epochs=2500)
    # test
    N = np.arange(1,train_loss.shape[0]+1)
    plt.plot(N, train_loss, label='eta = '+str(eta))

plt.xlabel("epochs")
plt.ylabel("training loss")
plt.title("exponential function - varying learning rates")
plt.legend()
plt.savefig('exponential function - learning rates .jpeg', dpi=800)
plt.show()
```

```python
x_train = np.expand_dims(np.linspace(0,2*np.pi,1000),axis=0) #-1 to 0 and 0.75 to 1.5
#x_test  =  #0 to 0.75

m = x_train.shape[1]
np.random.seed(48)
y_train = np.sin(x_train) + 0.02*np.random.randn(1,m)

eta_list = [0.05, 0.1, 0.2, 0.3]
nN = len(eta_list)
for i in range(nN):
    eta = eta_list[i]
    fcnet = Network()
    fcnet.add(fcLayer(1, 10,'tanh'))
    fcnet.add(regLayer(10, 1,'linear'))
    fcnet.optimizer(eta=eta, lamda=10, regularizer ='none') #L1
    # train
    [train_loss, test_loss] = fcnet.fit(x_train, y_train, x_train, y_train, batch_size = 8, epochs=2500)
    # test
    N = np.arange(1,train_loss.shape[0]+1)
    plt.plot(N, train_loss, label='eta = '+str(eta))

plt.xlabel("epochs")
plt.ylabel("training loss")
plt.title("sinusoidal function - varying learning rates")
plt.legend()
plt.savefig('sinusoidal function - learning rates .jpeg', dpi=800)
plt.show()
```
