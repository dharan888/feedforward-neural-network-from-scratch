```python
import numpy as np
import matplotlib.pyplot as plt

from neural_network import Network
from layers import fcLayer, classLayer, regLayer 

np.random.seed(24)

x_train = np.expand_dims(np.linspace(0,1,50),axis=0)
y_train = x_train + 0.04*np.random.randn(1,x_train.shape[1])
yhat = x_train
plt.rcParams['figure.figsize'] = [10, 6]
plt.scatter(x_train.T, y_train.T,s=15, facecolors='b', edgecolors='none', label='noised training data')
plt.plot(x_train.T, yhat.T,'y--',label='function to fit')

reg_list = ['none']

nN = len(reg_list)
for i in range(1):
    reg = reg_list[0]
    fcnet = Network()
    fcnet.add(fcLayer(1, 2,'linear'))
    fcnet.add(regLayer(2, 1,'linear'))
    fcnet.optimizer(eta=0.001, lamda=0.1, regularizer = reg) 
    # train
    [train_loss, test_loss] = fcnet.fit(x_train, y_train, x_train, y_train, batch_size = 1, epochs=500)
    y_pred = fcnet.predict(x_train)
    #predict
    plt.scatter(x_train, y_pred, s=15, facecolors='none', edgecolors='r', label="predicted")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Fitting linear function")
plt.legend()
plt.savefig('fitting linear function.jpeg', dpi=800)
plt.show()
```
