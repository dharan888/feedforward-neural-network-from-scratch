```python
import numpy as np
import matplotlib.pyplot as plt

from neural_network import Network
from layers import fcLayer, classLayer, regLayer 

np.random.seed(24)

x_train = np.expand_dims(np.linspace(-1,1,20),axis=0)
y_train = x_train + 0.1*np.random.randn(1,x_train.shape[1])
yhat = x_train

plt.rcParams['figure.figsize'] = [10, 6]
plt.scatter(x_train.T, y_train.T,s=15, facecolors='b', edgecolors='none', label='noised training data')
plt.plot(x_train.T, yhat.T,'y--',label='function to fit')

reg_list = ['none','L1','L2']

nN = len(reg_list)
for i in range(nN):
    reg = reg_list[i]
    fcnet = Network()
    fcnet.add(fcLayer(1, 100,'tanh'))
    fcnet.add(regLayer(100, 1,'linear'))
    fcnet.optimizer(eta=0.1, lamda=0.1, regularizer = reg) 
    # train
    [train_loss, test_loss] = fcnet.fit(x_train, y_train, x_train, y_train, batch_size = 4, epochs=10000)
    y_pred = fcnet.predict(x_train)
    #predict
    plt.plot(x_train.T, y_pred.T, label="predicted:"+reg + " regularization")
plt.xlabel("x")
plt.ylabel("y")
plt.title("effect of regularization on fitting linear function")
plt.legend()
plt.savefig('regularizitation_linear_function.jpeg', dpi=800)
plt.show()
```
