```python
import numpy as np
import matplotlib.pyplot as plt

from neural_network import Network
from layers import fcLayer, classLayer, regLayer 

x_train = np.expand_dims(np.linspace(-1,1,1000),axis=0)
y_train = np.exp(x_train) + 0.02*np.random.randn(1,x_train.shape[1])

plt.plot(x_train.T, y_train.T, label='y_true')
epochs_list = [100, 500, 1000, 2500]
nN = len(epochs_list)
for i in range(nN):
    epochs = epochs_list[i]
    fcnet = Network()
    fcnet.add(fcLayer(1, 10,'sigmoid'))
    fcnet.add(regLayer(10, 1,'linear'))
    fcnet.optimizer(eta=0.2, lamda=10, regularizer ='none') #L1
    # train
    [train_loss, test_loss] = fcnet.fit(x_train, y_train, x_train, y_train, batch_size = 8, epochs=epochs)
    # test
    predictions = fcnet.predict(x_train)
    plt.plot(x_train.T, predictions.T, label=str(epochs)+'epochs')

plt.xlabel("x")
plt.ylabel("y")
plt.title("exponential function - varying epochs")
plt.legend()
plt.savefig('neurons_exponential_epochs.jpeg', dpi=800)
plt.show()
```

```python
x_train = np.expand_dims(np.linspace(0,2*np.pi,1000),axis=0) #-1 to 0 and 0.75 to 1.5
#x_test  =  #0 to 0.75

m = x_train.shape[1]
y_train = np.sin(x_train) + 0.02*np.random.randn(1,m)

plt.plot(x_train.T, y_train.T, label='y_true')
epochs_list = [100, 500, 1000, 2500]
nN = len(epochs_list)
for i in range(nN):
    epochs = epochs_list[i]
    fcnet = Network()
    fcnet.add(fcLayer(1, 10,'tanh'))
    fcnet.add(regLayer(10, 1,'linear'))
    fcnet.optimizer(eta=0.2, lamda=10, regularizer ='none') #L1
    # train
    [train_loss, test_loss] = fcnet.fit(x_train, y_train, x_train, y_train, batch_size = 8, epochs=epochs)
    # test
    predictions = fcnet.predict(x_train)
    plt.plot(x_train.T, predictions.T, label=str(epochs)+'epochs')

plt.xlabel("x")
plt.ylabel("y")
plt.title("sinusoidal function - varying epochs")
plt.legend()
plt.savefig('neurons_sinusoidal_epochs.jpeg', dpi=800)
plt.show()
```
